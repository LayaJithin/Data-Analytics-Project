```python
# 1.Data Loading
import pandas as pd
student_data=pd.read_csv("https://github.com/user-attachments/files/22693045/Student.Performance.Factors.csv")
student_data

```

```python
# Number of rows and column in the dataset
print("The number of rows and columns in the dataset are :",student_data.shape)
```

```python
# Initial Overview
print("The First 5 rows of the Student Performance Dataset are: \n",student_data.head(5))
print("\n The Last 5 rows of the Student Performance Dataset are:\n",student_data.tail(5))
```

```python
# DataTypes of the Dataset
student_data.info()
```

```python
#Descriptive Summary of the Student Performance Dataset
student_data.describe()
```

```python
# Data Cleaning and Preprocessing
# Finding the missing values
missing_values=student_data.isnull().sum()
print("The number of missing values i each column\n",missing_values)
```

```python
# Imputing the null values
student_data['Teacher_Quality']=student_data['Teacher_Quality'].fillna(student_data['Teacher_Quality'].mode()[0])
student_data['Parental_Education_Level']=student_data['Parental_Education_Level'].fillna(student_data['Parental_Education_Level'].mode()[0])
student_data['Distance_from_Home']=student_data['Distance_from_Home'].fillna(student_data['Distance_from_Home'].mode()[0])

missing_values=student_data.isnull().sum()
missing_values
```

```python
# Remove the Duplicate rows in the dataset.
student_data=student_data.drop_duplicates()
student_data
```

```python
# Exploratory Data Analysis
# Count of each categorical Factors in the Student's Performance

print("Count of Parental Involvement in the Student Perfomance Factor Dataset :\n ",student_data['Parental_Involvement'].value_counts())
print("Count of Gender in the Student Perfomance Factor Dataset : \n ",student_data['Gender'].value_counts())
print("Count of Parental Education Level in the Student Perfomance Factor Dataset : \n",student_data['Parental_Education_Level'].value_counts())
print("Count of Teacher Quality in the Student Perfomance Factor Dataset : \n",student_data['Teacher_Quality'].value_counts())
print("Count of Distance from Home in the Student Perfomance Factor Dataset : \n",student_data['Distance_from_Home'].value_counts())
print("Count of Access to resources in the Student Perfomance Factor Dataset :\n ",student_data['Access_to_Resources'].value_counts())
print("Count of Motivation Level in the Student Perfomance Factor Dataset : \n",student_data['Motivation_Level'].value_counts())
print("Count of Teacher Quality in the Student Perfomance Factor Dataset : \n",student_data['Teacher_Quality'].value_counts())
print("Count of Peer influence in the Student Perfomance Factor Dataset : \n",student_data['Peer_Influence'].value_counts())
```

```python
# Unique values in the Parental Education Level,Peer_Influence,School Type
print("Unique values in the Parental Education Level : \n",student_data['Parental_Education_Level'].unique())
print("Unique values in the School Type : \n",student_data['School_Type'].unique())
print("Unique values in the Peer Influence : \n",student_data['Peer_Influence'].unique())
```

```python
# Filter columns by categorical values
categorical_columns=student_data.select_dtypes(include='object')
categorical_columns
```

```python
# Filtering column by numerical values
numerical_columns=student_data.select_dtypes(include='number')
numerical_columns
```

```python
# Comparison of Exam Score with Previous Score and its status

student_data['Score_Difference']=student_data['Exam_Score'].sub(student_data['Previous_Scores'])

student_data['Improvement_Status'] = student_data['Score_Difference'].apply(lambda x: 'Improved' if x > 1 else 'Not Improved')
student_data['Improvement_Status']

```

```python
# Visualization
# Visualisation showing the level of parental involvement
import seaborn as sn
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.subplot(2,2,1)
sn.countplot(x='Parental_Involvement',data=student_data)
```

```python
# visualization showing the count of male and female students in the dataset.
sn.countplot(x='Gender',data=student_data)
```

```python
# Scatter plot showing the Tutoring Session and Physical activity
plt.figure(figsize=(10,6))
sn.scatterplot(x='Physical_Activity',y='Tutoring_Sessions',data=student_data,hue='Gender')
plt.xlabel("Physical Activity")
plt.ylabel("Tutoring Sessions")
plt.title("Time Spent on Tutoring Sessions and Physical Activity by Male and female students")

```

```python
# barplot showing the Hours studied and Exam Score
plt.figure(figsize=(20,10))
sn.barplot(x='Exam_Score',y='Hours_Studied',data=student_data,hue="Gender")
plt.xlabel("Exam Score")
plt.ylabel("Hours Studied")
plt.title("Gender wise Hours studied vs Exam Score")
```

```python
# barplot showing Parental Involvement vs Exam score
plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
sn.barplot(x='Parental_Involvement',y='Score_Difference',data=student_data,hue='Gender')
plt.xlabel("Score Difference")
plt.ylabel("Parental Involvement")
plt.title("Gender wise -Parental Involvement Affecting Score ")
```

```python
# Piechart showing the Peer influence
plt.figure(figsize=(10,6))
plt.pie(student_data['Peer_Influence'].value_counts(),labels=student_data['Peer_Influence'].unique(),autopct='%1.1f%%')
plt.title("Peer Influence")
```

```python
# Pie chart showing Teacher Quality
plt.figure(figsize=(10,6))
plt.pie(student_data['Teacher_Quality'].value_counts(),labels=student_data['Teacher_Quality'].unique(),autopct='%1.1f%%')
plt.title("Teacher Quality")
```

```python
# Barplot showing the influence of internet access on Score
plt.figure(figsize=(10,6))
sn.barplot(x='Internet_Access',y='Score_Difference',data=student_data)
plt.xlabel("Internet Access")
plt.ylabel("Score Difference")
plt.title("Internet Access Vs Score Difference")
```

```python
# Barplot showing the average exam score vs Motivation_Level
plt.figure(figsize=(10,6))
order=(student_data.groupby('Motivation_Level')['Exam_Score'].mean().index)
sn.barplot(data=student_data,x='Motivation_Level',y='Exam_Score',order=order)
plt.xlabel("Motivation_Level")
plt.ylabel("Exam Score")
```

```python
# Boxplot showing Family Income vs Exam Score
plt.figure(figsize=(10,6))
sn.boxplot(x='Family_Income',y='Exam_Score',data=student_data)
plt.xlabel("Family Income")
plt.ylabel("Exam Score")
plt.title("Family Income Vs Exam Score")
```

```python
# Correlation between Exam Score,Physical Activity,Attendance,Hours Studied,Tutoring_Sessions,Sleep_Hours
plt.figure(figsize=(10,6))
corr_matrix=student_data[['Exam_Score','Physical_Activity','Attendance','Hours_Studied','Tutoring_Sessions','Sleep_Hours']].corr()

sn.heatmap(corr_matrix,cmap='coolwarm',annot=True)
plt.title("Correlation between Exam Score,Physical Activity,Attendance,Hours Studied,Tutoring_Sessions and Sleep_Hours")
```
